```python
from Crypto.Cipher import AES
from Crypto.Random import get_random_bytes
import time
```

```python
BLOCK_SIZE = AES.block_size

def pkcs7_pad(data: bytes) -> bytes:
    pad_len = BLOCK_SIZE - (len(data) % BLOCK_SIZE)
    return data + bytes([pad_len]) * pad_len

def pkcs7_unpad(data: bytes) -> bytes:
    pad_len = data[-1]
    return data[:-pad_len]

def aes_generate_key():
    return get_random_bytes(32)

def aes_encrypt(plaintext: bytes, key: bytes):
    iv = get_random_bytes(BLOCK_SIZE)
    cipher = AES.new(key, AES.MODE_CBC, iv)
    ciphertext = cipher.encrypt(pkcs7_pad(plaintext))
    return iv + ciphertext

def aes_decrypt(data: bytes, key: bytes):
    iv = data[:BLOCK_SIZE]
    ciphertext = data[BLOCK_SIZE:]
    cipher = AES.new(key, AES.MODE_CBC, iv)
    plaintext = cipher.decrypt(ciphertext)
    return pkcs7_unpad(plaintext)

key = aes_generate_key()
plaintext = b"Coba ganti saja kalimat ini dengan kalimatmu!"

enc = aes_encrypt(plaintext, key)
dec = aes_decrypt(enc, key)

print("Ciphertext length:", len(enc))
print("Recovered plaintext:", dec)
```
