```python
from Crypto.PublicKey import RSA
from Crypto.Cipher import PKCS1_OAEP
import time
```

```python
def rsa_generate_key(bits=2048):
    start = time.time()
    key = RSA.generate(bits)
    end = time.time()
    return key, end - start

def rsa_encrypt(message: bytes, public_key):
    cipher = PKCS1_OAEP.new(public_key)
    return cipher.encrypt(message)

def rsa_decrypt(ciphertext: bytes, private_key):
    cipher = PKCS1_OAEP.new(private_key)
    return cipher.decrypt(ciphertext)


# --- DEMO ---
key, t = rsa_generate_key(2048)
print("Generated RSA 2048 in:", t, "seconds")

public_key = key.publickey()

message = b"Coba ganti saja kalimat ini dengan kalimatmu"
ciphertext = rsa_encrypt(message, public_key)
plaintext = rsa_decrypt(ciphertext, key)

print("Ciphertext bytes:", len(ciphertext))
print("Recovered message:", plaintext)
```
