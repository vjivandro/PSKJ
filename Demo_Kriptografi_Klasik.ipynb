# **Teknik Subtitusi**

**1. Caesar Cipher**


```python
alphabet = "abcdefghijklmnopqrstuvwxyz"
partialOne=""
partialTwo=""
newAlphabet=""

message = input("Masukkan pesan (Plaintext) yang ingin dienkripsi: ").lower()  #
key = int(input("Masukkan nilai pergeseran (shift key, misalnya 3): ")) #

if key == 0:
 newAlphabet = alphabet
elif key > 0:
 partialOne = alphabet[:key]
 partialTwo = alphabet[key:]
 newAlphabet = partialTwo + partialOne
else:
 partialOne = alphabet[:(26 + key)]
 partialTwo = alphabet[(26 + key):]
 newAlphabet = partialTwo + partialOne

encrypted=""
for message_index in range(0,len(message)):
 if message[message_index] == " ":
  encrypted+= " "
  continue

 for alphabet_index in range(0,len(newAlphabet)):
  if message[message_index] == alphabet[alphabet_index]:
   encrypted+= newAlphabet[alphabet_index]
   break

decrypted = ""
for char in encrypted:
    if char == " ":
        decrypted += " "
        continue
    for alphabet_index in range(0, len(newAlphabet)):
        if char == newAlphabet[alphabet_index]:
            decrypted += alphabet[alphabet_index]
            break

print("\nPlaintext Asli:", message)
print("Ciphertext (Hasil Enkripsi):", encrypted)
print("\n")
print("Hasil Dekripsi:", decrypted)
```

**2. Mono-Alphabetic Cipher**

```python
SUBSTITUTION_KEY = {
    'a': 'q', 'b': 'w', 'c': 'e', 'd': 'r', 'e': 't',
    'f': 'y', 'g': 'u', 'h': 'i', 'i': 'o', 'j': 'p',
    'k': 'a', 'l': 's', 'm': 'd', 'n': 'f', 'o': 'g',
    'p': 'h', 'q': 'j', 'r': 'k', 's': 'l', 't': 'z',
    'u': 'x', 'v': 'c', 'w': 'v', 'x': 'b', 'y': 'n',
    'z': 'm', ' ': ' '
}

DECRYPTION_KEY = {v: k for k, v in SUBSTITUTION_KEY.items()}

def encrypt(plaintext, key):
    """Mengenkripsi plaintext menggunakan kunci substitusi mono-alfabetik."""
    ciphertext = ""
    plaintext = plaintext.lower()

    for char in plaintext:
        if char in key:
            ciphertext += key[char]
        else:
            ciphertext += char

    return ciphertext

def decrypt(ciphertext, key):
    """Mendekripsi ciphertext menggunakan kunci substitusi (kunci balik)."""
    plaintext = ""

    for char in ciphertext:
        if char in key:
            plaintext += key[char]
        else:
            plaintext += char

    return plaintext

plaintext_message = input("Masukkan pesan Plaintext: ").lower()

ciphertext_result = encrypt(plaintext_message, SUBSTITUTION_KEY)
print(f"Plaintext Asli: {plaintext_message}")
print(f"Ciphertext (Teks Sandi): {ciphertext_result}")
print("\n")
decrypted_result = decrypt(ciphertext_result, DECRYPTION_KEY)
print(f"Ciphertext yang Didekripsi: {ciphertext_result}")
print(f"Plaintext Hasil Dekripsi: {decrypted_result}")
```

# Teknik Transposisi

**1.Columnar Transposition (Abjad-Tunggal)**

```python
import math

def columnar_transposition_encrypt(plaintext, key):
    plaintext = plaintext.replace(" ", "").upper()
    num_cols = len(key)
    num_rows = math.ceil(len(plaintext) / num_cols)

    table = [['' for _ in range(num_cols)] for _ in range(num_rows)]
    index = 0
    for r in range(num_rows):
        for c in range(num_cols):
            if index < len(plaintext):
                table[r][c] = plaintext[index]
                index += 1

    key_order = sorted([(int(k), i) for i, k in enumerate(key)])
    ciphertext = ""
    for _, col_idx in key_order:
        for r in range(num_rows):
            if table[r][col_idx] != '':
                ciphertext += table[r][col_idx]

    return ciphertext


def columnar_transposition_decrypt(ciphertext, key):
    ciphertext = ciphertext.replace(" ", "").upper()
    num_cols = len(key)
    num_rows = math.ceil(len(ciphertext) / num_cols)

    table = [['' for _ in range(num_cols)] for _ in range(num_rows)]

    key_order = sorted([(int(k), i) for i, k in enumerate(key)])

    col_lengths = [num_rows] * num_cols
    total_cells = num_rows * num_cols
    extra_cells = total_cells - len(ciphertext)
    if extra_cells > 0:
        for i in range(extra_cells):
            col_lengths[key_order[-(i+1)][1]] -= 1

    index = 0
    for _, col_idx in key_order:
        for r in range(col_lengths[col_idx]):
            if index < len(ciphertext):
                table[r][col_idx] = ciphertext[index]
                index += 1

    plaintext = ""
    for r in range(num_rows):
        for c in range(num_cols):
            if table[r][c] != '':
                plaintext += table[r][c]

    return plaintext

print("COLUMNAR TRANSPOSITION CIPHER")
plaintext = input("Masukkan pesan asli (plaintext): ")
key = input("Masukkan kunci angka unik (misal: 3142 atau 431256): ")

ciphertext = columnar_transposition_encrypt(plaintext, key)
print("\nHasil Enkripsi:")
print(ciphertext)

decrypted = columnar_transposition_decrypt(ciphertext, key)
print("\nHasil Dekripsi:")
print(decrypted)
```

**2. Rail Fence Cipher**

```python
def encryptRailFence(text, key):
    """
    Fungsi enkripsi Rail Fence.
    """
    text = "".join(text.split()).lower()
    L = len(text)

    rail = [['\n' for _ in range(L)] for _ in range(key)] # [1]

    dir_down = False
    row, col = 0, 0

    for i in range(L):
        if (row == 0) or (row == key - 1):
            dir_down = not dir_down

        rail[row][col] = text[i]
        col += 1

        if dir_down:
            row += 1
        else:
            row -= 1
    result = []
    for i in range(key):
        for j in range(L):
            if rail[i][j]!= '\n':
                result.append(rail[i][j])

    return "".join(result)

def decryptRailFence(cipher, key):
    """
    Fungsi dekripsi Rail Fence.
    Memerlukan tiga fase: Menandai, Mengisi, dan Membaca ulang.
    """
    L = len(cipher)
    rail = [['\n' for _ in range(L)] for _ in range(key)]

    dir_down = False
    row, col = 0, 0
    for i in range(L):
        if (row == 0) or (row == key - 1):
            dir_down = not dir_down
        rail[row][col] = '*'
        col += 1

        if dir_down:
            row += 1
        else:
            row -= 1

    index = 0
    for i in range(key):
        for j in range(L):
            if rail[i][j] == '*' and index < L:
                rail[i][j] = cipher[index]
                index += 1

    result = []
    dir_down = False
    row, col = 0, 0

    for i in range(L):
        if (row == 0) or (row == key - 1):
            dir_down = not dir_down

        if rail[row][col]!= '*':
            result.append(rail[row][col])
        col += 1

        if dir_down:
            row += 1
        else:
            row -= 1

    return "".join(result)


print("="*50)
print("DEMO: RAIL FENCE CIPHER")
print("="*50)

plaintext_rf = input("Masukkan Plaintext untuk Rail Fence: ")
key_rf = 3

ciphertext_rf = encryptRailFence(plaintext_rf, key_rf)
print(f"Plaintext Asli: {plaintext_rf}")
print(f"Kunci (Jumlah Rel): {key_rf}")
print(f"Ciphertext: {ciphertext_rf}")

decrypted_rf = decryptRailFence(ciphertext_rf, key_rf)
print(f"Hasil Dekripsi: {decrypted_rf}")
print("="*50)
```

**3. Route Cipher**

```python
def encryptSpiral(text, width):
    """
    Fungsi enkripsi Sandi Spiral (Out Spirally).
    Kunci 'width' menentukan lebar matriks.
    """
    text = "".join(text.split()).lower()
    L = len(text)

    height = (L + width - 1) // width
    padding_length = (width * height) - L

    padded_text = text + ('-' * padding_length)

    matrix = []
    for i in range(height):
        start = i * width
        end = (i + 1) * width
        matrix.append(list(padded_text[start:end]))

    ciphertext = ""
    top, bottom = 0, height - 1
    left, right = 0, width - 1

    while top <= bottom and left <= right:
        for i in range(left, right + 1):
            ciphertext += matrix[top][i]
        top += 1
        for i in range(top, bottom + 1):
            ciphertext += matrix[i][right]
        right -= 1

        if top <= bottom:
            for i in range(right, left - 1, -1):
                ciphertext += matrix[bottom][i]
            bottom -= 1

        if left <= right:
            for i in range(bottom, top - 1, -1):
                ciphertext += matrix[i][left]
            left += 1

    return ciphertext.replace('-', '')

def decryptSpiral(cipher, key_width):
    """
    Fungsi dekripsi Sandi Spiral.
    Membalikkan proses Traversal Spiral untuk mengisi matriks.
    """
    L = len(cipher)
    width = key_width
    height = (L + width - 1) // width
    padding_length = (width * height) - L

    cipher_padded = cipher + (' ' * padding_length)

    matrix = [['' for _ in range(width)] for _ in range(height)]

    top, bottom = 0, height - 1
    left, right = 0, width - 1
    char_index = 0

    while top <= bottom and left <= right:
        for i in range(left, right + 1):
            matrix[top][i] = cipher_padded[char_index]
            char_index += 1
        top += 1
        for i in range(top, bottom + 1):
            matrix[i][right] = cipher_padded[char_index]
            char_index += 1
        right -= 1
        if top <= bottom:
            for i in range(right, left - 1, -1):
                matrix[bottom][i] = cipher_padded[char_index]
                char_index += 1
            bottom -= 1
        if left <= right:
            for i in range(bottom, top - 1, -1):
                matrix[i][left] = cipher_padded[char_index]
                char_index += 1
            left += 1

    plaintext_result = ""
    for row in matrix:
        plaintext_result += "".join(row)
    return plaintext_result.replace(' ', '')

print("DEMO: SPIRAL CIPHER")
print("="*50)
plaintext_spiral = input("Masukkan Plaintext untuk Spiral Cipher: ")
key_width_spiral = 6

ciphertext_spiral = encryptSpiral(plaintext_spiral, key_width_spiral)
print(f"Plaintext Asli: {plaintext_spiral}")
print(f"Kunci (Lebar Matriks): {key_width_spiral}")
print(f"Ciphertext: {ciphertext_spiral}")

decrypted_spiral = decryptSpiral(ciphertext_spiral, key_width_spiral)
print(f"Hasil Dekripsi: {decrypted_spiral}")
print("="*50)
```
